* Dodaj kolumnę i policz wiek aktorów 
* Usuń kolumny (bio, death_details)
* Zmień nazwy kolumn - dodaj kapitalizaję i usuń _

```python

from pyspark.sql.functions import current_timestamp, unix_timestamp, col, count, year, when, current_date, to_date, regexp_extract, abs
filePath = "dbfs:/FileStore/tables/Files/names.csv"
namesDf = spark.read.format("csv").option("header", "true").option("inferSchema", "true").load(filePath)
namesDf = namesDf.withColumn("ExecutionTime", unix_timestamp(current_timestamp()))
namesDf = namesDf.withColumn("feet", col("height") * 0.0328084)
name = namesDf.groupBy("name").agg(count("*").alias("count")).orderBy(col("count").desc()).limit(10)
name.explain()
display(name)
# David Campbell i David Bradley 
namesDf = namesDf.drop("bio", "death_details")
new_column_names = [col.replace("_", "") for col in namesDf.columns]
namesDf = namesDf.toDF(*new_column_names)
namesDf = namesDf.orderBy("Name")
namesDf = namesDf.withColumn("ExecutionTime", abs(unix_timestamp(current_timestamp()) - col("ExecutionTime")))
namesDf.explain()
display(namesDf)
grouped_df = namesDf.groupBy("children").agg({"divorces": "avg"})
display(grouped_df)
namesDf.explain()
display(namesDf)
```

ratings.csv
* Dodaj kolumnę z wartością czasu wykonania notatnika w formacie epoch
* Dla każdego z poniższych wyliczeń nie bierz pod uwagę `nulls` 
* Kto daje lepsze oceny chłopcy czy dziewczyny dla całego setu
* Dla jednej z kolumn zmień typ danych do `long` 

```python

filePath = "dbfs:/FileStore/tables/Files/ratings.csv"
namesDf = spark.read.format("csv").option("header","true").option("inferSchema","true").load(filePath)

from pyspark.sql.functions import unix_timestamp, current_timestamp, col, avg

ratingsDf = namesDf.withColumn("execution_time_epoch", unix_timestamp(current_timestamp()))

ratingsNoNulls = ratingsDf.dropna(subset=["males_allages_avg_vote", "females_allages_avg_vote"])

male_avg = ratingsNoNulls.select(avg("males_allages_avg_vote").alias("male_avg")).first()["male_avg"]
female_avg = ratingsNoNulls.select(avg("females_allages_avg_vote").alias("female_avg")).first()["female_avg"]

if male_avg > female_avg:
    print(f"Chłopcy dają wyższe oceny (średnia: {male_avg:.2f} vs {female_avg:.2f})")
elif female_avg > male_avg:
    print(f"Dziewczyny dają wyższe oceny (średnia: {female_avg:.2f} vs {male_avg:.2f})")
else:
    print(f"Oceny chłopców i dziewczyn są równe (średnia: {male_avg:.2f})")

ratingsDf = ratingsDf.withColumn("total_votes_long", col("total_votes").cast("long"))

ratingsDf.select("imdb_title_id", "total_votes", "total_votes_long", "execution_time_epoch").show(5)


#Spark UI pozwala na monitorowanie działania aplikacji Spark i analizę jej wydajności. W zakładce Stages można zobaczyć, jak zadania zostały podzielone na mniejsze etapy oraz ile czasu zajęło ich wykonanie. Jobs zawiera listę uruchomionych zadań wraz z ich statusami, co ułatwia śledzenie postępu pracy.

#Zakładka Storage umożliwia sprawdzenie, czy dane są przechowywane w pamięci oraz ile miejsca zajmują, co jest kluczowe dla optymalizacji wydajności. Environment dostarcza informacji o konfiguracji środowiska, a Executors przedstawia statystyki związane z działaniem poszczególnych wykonawców. Dodatkowo, w SQL można znaleźć szczegóły dotyczące wykonywanych zapytań, co ułatwia analizę ich efektywności.

#Dzięki Spark UI można nie tylko śledzić postęp obliczeń, ale także optymalizować działanie aplikacji i rozwiązywać potencjalne problemy. 
```
